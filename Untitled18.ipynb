```python
while True:
    print("Bem vindo à Calculadora Inteligente")
    print("Escolha uma operação do 1 ao 6")
    print("1. Soma")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")
    print("5. Potenciação")
    print("6. Sair")

    número = int(input("Digite o número da operação: "))

    if número == 6:
        print("Fim do código")
        break

    if número < 1 or número > 6:
        print("Opção inválida.")

    num1 = int(input("Digite o primeiro número: "))
    num2 = int(input("Digite o segundo número: "))

    if número == 1:
        print(f"Soma: {num1 + num2}")
    elif número == 2:
        print(f"Subtração: {num1 - num2}")
    elif número == 3:
        print(f"Multiplicação: {num1 * num2}")
    elif número == 4:
         if num2 == 0:
          print("Erro: Não é possível dividir por zero.")
         else:
            print(f"Divisão: {num1 / num2}")
    elif número == 5:
        print(f"Potenciação: {num1 ** num2}")
```
